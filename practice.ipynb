```python
# ১. প্রয়োজনীয় টুলবক্স বা লাইব্রেরি ইমপোর্ট করা
import csv

# ২. খরচ যোগ করার জন্য একটি ফাংশন বানানো
# def মানে হলো আমরা একটি নতুন কাজ বা ফাংশন ডিফাইন করছি
def add_expense(filename):
    print("\n--- নতুন খরচ যোগ করুন ---")
    
    # ইনপুট নেওয়া: এখানে item একটি স্ট্রিং এবং amount একটি দশমিক সংখ্যা (float)
    item = input("কী কিনলেন? : ")
    amount = float(input("কত টাকা? : ")) 
    category = input("বিভাগ (খাবার/ভ্রমণ/শপিং): ")
    
    # একটি ডিকশনারিতে তথ্যগুলো গুছিয়ে রাখা
    expense = {
        "Item": item,
        "Amount": amount,
        "Category": category
    }
    
    # 'with open' ব্যবহার করা যাতে কাজ শেষে ফাইল অটোমেটিক বন্ধ হয়ে যায়
    # 'a' (append) মোড মানে হলো পুরনো ডেটা মুছবে না, নিচে নিচে নতুন ডেটা যোগ হবে
    with open(filename, mode='a', newline='') as file:
        # কলামের নামগুলো বলে দেওয়া
        fieldnames = ["Item", "Amount", "Category"]
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        
        # ফাইলটি যদি একদম নতুন হয় (খালি থাকে), তবে হেডার বা কলামের নাম লিখে দেওয়া
        file.seek(0, 2) # চেক করা ফাইলের ভেতর কিছু আছে কি না
        if file.tell() == 0:
            writer.writeheader()
            
        # আসল ডেটাটি ফাইলে লিখে ফেলা
        writer.writerow(expense)
    print("✅ খরচটি সফলভাবে সেভ হয়েছে।")

# ৩. জমানো খরচগুলো দেখার জন্য ফাংশন
def show_report(filename):
    print("\n--- খরচের রিপোর্ট ---")
    total = 0 # মোট খরচের হিসাব রাখার জন্য একটি ভ্যারিয়েবল
    
    # try...except ব্যবহার করা যাতে ফাইল না থাকলে প্রোগ্রাম ক্র্যাশ না করে
    try:
        with open(filename, mode='r') as file:
            reader = csv.DictReader(file)
            # লুপ চালিয়ে এক লাইন এক লাইন করে পড়া
            for row in reader:
                print(f"{row['Item']} - {row['Amount']} টাকা ({row['Category']})")
                total += float(row['Amount']) # স্ট্রিং থেকে নাম্বারে নিয়ে যোগ করা
        print(f"-------------------")
        print(f"মোট খরচ: {total} টাকা")
    except FileNotFoundError:
        print("❌ কোনো ফাইল পাওয়া যায়নি!")

# ৪. মেইন প্রোগ্রাম (যেখান থেকে সবকিছু শুরু হবে)
def main():
    filename = "expenses.csv"
    
    # একটি ইনফিনিট লুপ (যতক্ষণ না ইউজার ৩ টিপে বের হবে)
    while True:
        print("\n--- মেনু ---")
        print("১. খরচ যোগ করুন")
        print("২. রিপোর্ট দেখুন")
        print("৩. বের হয়ে যান")
        
        choice = input("আপনার পছন্দ (১/২/৩): ")
        
        # ইউজার কী ইনপুট দিচ্ছে তার ওপর ভিত্তি করে ডিসিশন নেওয়া
        if choice == '1':
            add_expense(filename) # যোগ করার ফাংশন কল করা
        elif choice == '2':
            show_report(filename) # রিপোর্ট দেখানোর ফাংশন কল করা
        elif choice == '3':
            print("বিদায়!")
            break # লুপ থেকে বের হয়ে আসা
        else:
            print("ভুল বাটন চেপেছেন!")

# এটি নিশ্চিত করে যে প্রোগ্রামটি সরাসরি রান করলে মেইন ফাংশনটি কাজ শুরু করবে
if __name__ == "__main__":
    main()
```
