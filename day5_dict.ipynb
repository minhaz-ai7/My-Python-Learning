```python
students = [
    {"name": "Minhaz", "age": 21, "marks": 232},
    {"name": "Rahim", "age": 22, "marks": 567},
    {"name": "Karim", "age": 20, "marks": 500}
]

total = 0

for x in students:
    total += x['age']

average = total / len(students) 

print(average)
```
