```python
name = input("whats your name: ").title()

if name == "Harry" or name == "Rahman" or name == "Ron":
    print("Gryffindor")

elif name == "Draco":
    print("Slytherin")

else:
    print("who.? ")    
```

```python
name = input("whats your name: ").title()

match name:
    case "harry" | "rahman" | "ron":
        print("gryffinder")

    case _:
        print("who.?")
```
